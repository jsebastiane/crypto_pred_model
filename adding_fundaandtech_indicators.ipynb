#### Creating RSI and Moving average
###### Using 14 trailing days
###### RSI = 100 - 100/ (1 + RS)
###### RS = [(previous average gain) * 13 + current gain)/14] / [(previous average loss) * 13 + current loss)/14]

```python
# Getting First Average Gain and Loss in order to get first RSI
list_closes = eth_data['close']
sum_losses = 0
sum_gains = 0
for i in range(1,15):
    change = ((list_closes[i] - list_closes[i-1])/list_closes[i-1]) * 100
    if change < 0 :
        sum_losses += change * -1
    else:
        sum_gains += change


first_average_gain = sum_gains/14
first_average_loss = sum_losses/14



first_rsi = 100 - (100/(1+(first_average_gain/first_average_loss)))

# (first_average_gain * 13 + current_gain) /14 / (first_average_loss * 13 + current_loss)/14

    
    
    
```

```python
# Keeping track of previous average gain and average loss
eth_data['rsi'] = np.nan
eth_data['cag'] = np.nan
eth_data['cal'] = np.nan
eth_data.at[14, 'rsi'] = first_rsi
eth_data.at[14, 'cag'] = first_average_gain
eth_data.at[14, 'cal'] = first_average_loss
```

```python
print(eth_data.loc[14])
```

```python
# Function to calculate RSI
def calc_rsi(cag, cal, close, prev_close):
    change = ((close - prev_close)/prev_close) * 100
    new_cag = 0
    new_cal = 0
    if change < 0:
        new_cag = ((cag * 13) + 0)/14
        new_cal = ((cal * 13) + (change * -1))/14
    else:
        new_cag = ((cag * 13) + change)/14
        new_cal = ((cal * 13) + 0)/14
        
    new_rsi = 100 - (100/(1 + (cag/cal)))
    
    return new_cag, new_cal, new_rsi
    
```

```python
cag, cal, rsi = calc_rsi(3,4, 1345, 1356)
print(cag)
print(cal)
print(rsi)
```

```python
for i in range(15, len(eth_data)):
    cag = eth_data.at[i-1, 'cag']
    cal = eth_data.at[i-1, 'cal']
    close = eth_data.at[i, 'close']
    prev_close = eth_data.at[i-1, 'close']
    new_cag, new_cal, new_rsi = calc_rsi(cag, cal, close, prev_close)
    eth_data.at[i, 'rsi'] = new_rsi
    eth_data.at[i, 'cag'] = new_cag
    eth_data.at[i, 'cal'] = new_cal
```

```python
# Checkpoint
eth_data.to_csv('eth_daily_main.csv')
```

```python
# EMA Calculation
#  multiplier = (2 * [#of observations + 1])
# EMA = CLOSE * multiplier + EMA(i-1)*(1-multiplier)
eth_data['ema'] = np.nan

# Step 1 - calculate SMA (first EMA) of first 14 days
list_closes = eth_data['close']
# Here 1:14 includes 14 - it's not like range where (1,14) does not include 14
sum_closes = eth_data.loc[1:14,['close']].sum()
eth_data.at[14, 'ema'] = sum_closes/14
```

```python
# Step 2 - create function to calc EMA
def cal_ema(close, prev_ema):
    multiplier = 2 / (14 + 1)
    ema = (close * multiplier) + (prev_ema * (1-multiplier))
    return ema
    
```

```python
# Step 3 - call calc_ema function to isnert into table
for i in range(15, len(eth_data)):
    close = eth_data.at[i, 'close']
    prev_ema = eth_data.at[i-1, 'ema']
    ema = cal_ema(close, prev_ema)
    eth_data.at[i, 'ema'] = ema
```

```python
# Checkpoint 2
eth_data.to_csv('eth_daily_main')
```
