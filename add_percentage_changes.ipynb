```python
import pandas as pd
```

```python
df = pd.read_csv('eth_daily.csv')
df
```

```python
# iterate through all closing prices
# calculate different between prices in percentage
# data = df.iloc[0]["close"]
data = df.reindex(index = df.index[::-1])
temp_list = []
for i in range(0,2127):
    temp_list.append(i)
    
data["index"] = temp_list
```

```python
data
```

```python
data.set_index('index', inplace=True)
data
```

```python
closing_prices = data.iloc[0:len(data.index)]['close']  
closing_prices
    
    
```

```python
percentage_changes = []
for i in range(1,len(closing_prices)):
    quotient = (closing_prices[i]-closing_prices[i-1])/closing_prices[i-1]
    percentage_change = quotient*100
    percentage_changes.append(percentage_change)
    
```

```python
percentage_changes.insert(0,0)
```

```python
data["prcntg_chg"] = percentage_changes
```

```python
data
```

```python
data.to_csv('eth_daily_edited.csv', index=False)
```
